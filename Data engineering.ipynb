```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime
import missingno as msno
```

```python
df = pd.read_csv('./input/wind_dataset.csv')
df.info()
```

```python
df['DATE']=pd.to_datetime(df['DATE'])
df = df.set_index('DATE')
dataset = df.copy()
df.head()
```

## T.MAX

```python
df.plot(y='T.MAX',figsize=(15,8))
```

```python
df = df.assign(FillMean=df['T.MAX'].fillna(df['T.MAX'].mean()))
df = df.assign(FillMedian=df['T.MAX'].fillna(df['T.MAX'].median()))
df = df.assign(Ffill=df['T.MAX'].fillna(df['T.MAX'].ffill()))
df = df.assign(Bfill=df['T.MAX'].fillna(df['T.MAX'].bfill()))
df = df.assign(InterpolateLinear=df['T.MAX'].interpolate(method='linear'))
df = df.assign(InterpolateTime=df['T.MAX'].interpolate(method='time'))
df = df.assign(InterpolateQuadratic=df['T.MAX'].interpolate(method='quadratic'))
df = df.assign(InterpolateCubic=df['T.MAX'].interpolate(method='cubic'))
df = df.assign(InterpolateSLinear=df['T.MAX'].interpolate(method='slinear'))
df = df.assign(InterpolateAkima=df['T.MAX'].interpolate(method='akima'))
df = df.assign(InterpolatePoly5=df['T.MAX'].interpolate(method='polynomial', order=5)) 
df = df.assign(InterpolatePoly7=df['T.MAX'].interpolate(method='polynomial', order=7))
df = df.assign(InterpolateSpline3=df['T.MAX'].interpolate(method='spline', order=3))
df = df.assign(InterpolateSpline4=df['T.MAX'].interpolate(method='spline', order=4))
```

```python
df.plot(y='FillMean',figsize=(15,8))
df.plot(y='FillMedian',figsize=(15,8))
df.plot(y='Ffill',figsize=(15,8))
df.plot(y='Bfill',figsize=(15,8))
df.plot(y='InterpolateLinear',figsize=(15,8))
df.plot(y='InterpolateTime',figsize=(15,8))
df.plot(y='InterpolateQuadratic',figsize=(15,8))
df.plot(y='InterpolateCubic',figsize=(15,8))
df.plot(y='InterpolateSLinear',figsize=(15,8))
df.plot(y='InterpolateAkima',figsize=(15,8))
df.plot(y='InterpolatePoly5',figsize=(15,8))
df.plot(y='InterpolatePoly7',figsize=(15,8))
df.plot(y='InterpolateSpline3',figsize=(15,8))
df.plot(y='InterpolateSpline4',figsize=(15,8))
```

```python
df['T.MAX'] = df['InterpolateLinear']
df.plot(y='T.MAX',figsize=(15,8))
```
